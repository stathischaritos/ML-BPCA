<h2>1. The $Q$-distribution</h2>
The function that $Q$ approximates is $P(\Theta|D)$:

We know:

$$\ln p(D) = \mathcal{L}(Q) + \mathrm{KL}(Q||P)$$

This can also be written as:

$$\ln p(D) = \mathcal{L}(Q) + \int{Q(\Theta)ln\frac{P(\Theta|D)}{Q(\Theta)}d_{\Theta} }$$
where the KL divergence (second term in the summation) describes the difference between the lower bound $\mathcal{L}(Q)$ and $ln P(D)$.
We know that the KL divergence is minimal (equal to zero) when the two distributions are the same.
Hence this error term will vanish, if the distribution $Q(\Theta)=P(\Theta|D)$. Thus from the given equation we can consider $Q(\Theta)$ as an approximation of $P(\Theta|D)$.

<h2>2. The mean-field approximation</h2>
Following the steps mentioned in the assignment (similar to section 10.1.1 in Bishop's text book), we have:

$$\mathcal{L}(Q) = \int{ Q(\Theta) \ln{\frac{p(D,\Theta)}{Q(\Theta)}} }d\Theta$$

We now use $Q(\Theta) = \prod_k{Q_k(\Theta_k)}$:

$$\mathcal{L}(Q) = \int{ \prod_k{Q_k(\Theta_k)} \ln{\frac{p(D,\Theta)}{\prod_k{Q_k(\Theta_k)}}} }d\Theta$$
$$= \int{ \prod_{k \neq i}{Q_k(\Theta_k)} Q_i(\Theta_i) [ \ln{p(D,\Theta) - \sum_k{\ln{Q_k(\Theta_k)}}} ] }d\Theta$$
$$= \int{ Q_i(\Theta_i) [ \int{ \prod_{k \neq i}{Q_k(\Theta_k)} \ln{p(D,\Theta)} }d\Theta_k ] }d\Theta_i 
- \int{ \prod_{k \neq i}{Q_k(\Theta_k)} Q_i(\Theta_i) \sum_k{\ln{Q_k(\Theta_k)}} }d\Theta$$
$$= \int{ Q_i(\Theta_i) [ \int{ \prod_{k \neq i}{Q_k(\Theta_k)} \ln{p(D,\Theta)} }d\Theta_k ] }d\Theta_i 
- \int{ Q_i(\Theta_i) [ \int{ \prod_{k \neq i}{Q_k(\Theta_k)} \sum_k{\ln{Q_k(\Theta_k)}} }d\Theta_k] }d\Theta_i$$
$$= \int{ Q_i(\Theta_i) [ \int{ \prod_{k \neq i}{Q_k(\Theta_k)} \ln{p(D,\Theta)} }d\Theta_k ] }d\Theta_i 
- \int{ Q_i(\Theta_i) [ \int{ \prod_{k \neq i}{Q_k(\Theta_k)} \sum_{k \neq i}{\ln{Q_k(\Theta_k)}} }d\Theta_k 
+ \int{ \prod_{k \neq i}{Q_k(\Theta_k)} \ln{Q_i(\Theta_i)} }d\Theta_k ] }d\Theta_i$$

We note here that $\int{ \prod_{k \neq i}{Q_k(\Theta_k)} \sum_{k \neq i}{\ln{Q_k(\Theta_k)}} }d\Theta_k$ is a constant. Thus,

$$\mathcal{L}(Q) = \int{ Q_i(\Theta_i) [ \int{ \prod_{k \neq i}{Q_k(\Theta_k)} \ln{p(D,\Theta)} }d\Theta_k ] }d\Theta_i 
- \int{ Q_i(\Theta_i) [ const + \ln{Q_i(\Theta_i)} \int{ \prod_{k \neq i}{Q_k(\Theta_k)} }d\Theta_k ] }d\Theta_i$$

This time we note that $\int{ \prod_{k \neq i}{Q_k(\Theta_k)} }d\Theta_k$ is equal to $1$. Therefore,

$$\mathcal{L}(Q) = \int{ Q_i(\Theta_i) [ \int{ \prod_{k \neq i}{Q_k(\Theta_k)} \ln{p(D,\Theta)} }d\Theta_k ] }d\Theta_i 
- \int{ Q_i(\Theta_i) [ const + \ln{Q_i(\Theta_i)} ] }d\Theta_i$$
$$= \int{ Q_i(\Theta_i) [ \int{ \prod_{k \neq i}{Q_k(\Theta_k)} \ln{p(D,\Theta)} }d\Theta_k ] }d\Theta_i 
- const \int{ Q_i(\Theta_i) }d\Theta_i - \int{ Q_i(\Theta_i) }d\Theta_i$$

Here, note that $\int{ Q_i(\Theta_i) }d\Theta_i$ is also equal to $1$. So,

$$\mathcal{L}(Q) = \int{ Q_i(\Theta_i) [ \int{ \prod_{k \neq i}{Q_k(\Theta_k)} \ln{p(D,\Theta)} }d\Theta_k ] }d\Theta_i 
- \int{ Q_i(\Theta_i) }d\Theta_i + const $$

We now define $\ln{\tilde{p}(D,\Theta)} = \mathbb{E}_{k \neq i}[ln{p(D,\Theta)}] + const$. Thus,

$$\mathcal{L}(Q) = \int{ Q_i(\Theta_i) \ln{\tilde{p}(D,\Theta)} }d\Theta - \int{ Q_i(\Theta_i) \ln{Q_i(\Theta_i)} }d\Theta $$
$$\mathcal{L}(Q) = \int{ Q_i(\Theta_i) \ln{\frac{\tilde{p}(D,\Theta)}{Q_i(\Theta_i)}} } + const $$

We recognize this as the negetive KL divergence between $Q_i(\Theta_i)$ and $\tilde{p}(D,\Theta)$.

This is maximized when the KL divergence is minimized, and the KL divergence is minimized when $Q_i(\Theta_i) = \tilde{p}(D,\Theta)$

Thus, an optimal solution would be:

$$ \ln{Q_i^*(\Theta_i)} = \mathbb{E}_{k \neq i}[ln{p(D,\Theta)}] + const$$

The equation following eq. 10.9 in Bishop's textbook contains the exponent and is normalized to eliminate the constant:

$$ Q_i^*(\Theta_i) = \frac{ \exp{[\mathbb{E}_{k \neq i}[ln{p(D,\Theta)}]]} } { \int{exp{[\mathbb{E}_{k \neq i}[ln{p(D,\Theta)}]]}}d\Theta } $$

NOTE: The subscript $j$ in eq. 13 of [Bishop99] should be $i$.

The entropy terms can easily be looked up and are given by:

<h3>Gamma distribution:</h3>
$$a-ln(b)+ln[\Gamma(a)] +(1-a)\psi(a)$$
where a and b are parameters of the gamma function
<h3>Normal distribution:</h3>
$$\frac{d}{2} (1 + ln(2\pi) + \frac{1}{2} ln|\Sigma|)$$
where d is the dimensionality of the normal distribution and $\Sigma$ the covariance.



dropping the const values in the evaluated integrals and given the entropies, we can compute the lower bound.
Also monitoring the change of the lower bound can be used as a measure for convergence (for which we do not need constants).

```python
%pylab inline

import numpy as np
import scipy.special as sp
from numpy.core.umath_tests import inner1d
import pylab as plt
import cPickle, gzip

#Plotting methods
#Note that these functions have been taken from http://wiki.scipy.org/Cookbook/Matplotlib/HintonDiagrams
#We do not claim any authorship to these functions. 
def _blob(x,y,area,colour):
    """
    Draws a square-shaped blob with the given area (< 1) at
    the given coordinates.
    """
    hs = np.sqrt(area) / 2
    xcorners = np.array([x - hs, x + hs, x + hs, x - hs])
    ycorners = np.array([y - hs, y - hs, y + hs, y + hs])
    plt.fill(xcorners, ycorners, colour, edgecolor=colour)

def hinton(W, maxWeight=None):
    """
    Draws a Hinton diagram for visualizing a weight matrix. 
    Temporarily disables matplotlib interactive mode if it is on, 
    otherwise this takes forever.
    """
    reenable = False
    if plt.isinteractive():
        plt.ioff()
    plt.clf()
    height, width = W.shape
    if not maxWeight:
        maxWeight = 2**np.ceil(np.log(np.max(np.abs(W)))/np.log(2))

    plt.fill(np.array([0,width,width,0]),np.array([0,0,height,height]),'gray')
    plt.axis('off')
    plt.axis('equal')
    for x in xrange(width):
        for y in xrange(height):
            _x = x+1
            _y = y+1
            w = W[y,x]
            if w > 0:
                _blob(_x - 0.5, height - _y + 0.5, min(1,w/maxWeight),'white')
            elif w < 0:
                _blob(_x - 0.5, height - _y + 0.5, min(1,-w/maxWeight),'black')
    if reenable:
        plt.ion()
    return plt.show()

#The Class
class BayesianPCA(object):
    
    def __init__(self, d, N, a_alpha=10e-3, b_alpha=10e-3, a_tau=10e-3, b_tau=10e-3, beta=10e-3):
        """
        """
        self.d = d # number of dimensions
        self.N = N # number of data points
        
        # Hyperparameters
        self.a_alpha = a_alpha
        self.b_alpha = b_alpha
        self.a_tau = a_tau
        self.b_tau = b_tau
        self.beta = beta

        # Variational parameters
        self.means_z = np.random.randn(d, N) # called x in bishop99
        self.sigma_z = np.random.randn(d, d)
        self.mean_mu = np.random.randn(d, 1)
        self.sigma_mu = np.random.randn(d, d)
        self.means_w = np.random.randn(d, d)
        self.sigma_w = np.random.randn(d, d)
        self.a_alpha_tilde = np.abs(np.random.randn(1))
        self.bs_alpha_tilde = np.abs(np.random.randn(d, 1))
        self.a_tau_tilde = np.abs(np.random.randn(1))
        self.b_tau_tilde = np.abs(np.random.randn(1))
    
    #computes expectation of tau
    def __exp_tau(self):
        return (self.a_tau_tilde / self.b_tau_tilde)
    
    # computes expectation of W_Transpose * W
    def __exp_w_T_w(self):
        wtw = np.dot(self.means_w.T, self.means_w)
        w_sig = np.sum(self.sigma_w, axis=0, keepdims=True)
        diag_w_sig = np.diag(w_sig)
        exp_wtw = np.add(diag_w_sig, wtw)
        return exp_wtw
    
    # compute expectation of W
    def __exp_W(self):
        return(self.means_w)
    
    # computes expectation of z_n
    def __exp_zn(self, n):
        return self.means_z[:,n]
    
    def __exp_z(self):
        return(self.means_z)
    
    #computes expectation of mu
    def __exp_mu(self):
        return self.mean_mu
        
    #computes expectation of alpha
    def __exp_alpha(self):
        exp_alpha = self.a_alpha_tilde / self.bs_alpha_tilde
        return exp_alpha
        
    def __exp_zn_znt(self,n):
        zn_znt = np.add(self.sigma_z, np.dot(self.means_z[:,n], self.means_z[:,n].T))
        return zn_znt
    
    def __exp_z_zt(self):
        z_zt = np.add(self.N * self.sigma_z, np.dot(self.means_z, np.transpose(self.means_z)))
        return z_zt
    
    def __update_z(self, X):
        #sigma_Z
        I = np.identity(self.d)
        self.sigma_z = np.linalg.inv(np.add(I, self.__exp_tau() * self.__exp_w_T_w()))
        
        #mean_Z
        wt = self.__exp_W().T
        self.means_z = self.__exp_tau() * np.dot(self.sigma_z, np.dot(wt, X - self.__exp_mu())) 

    def __update_mu(self, X):
        #Update sigma
        I=np.identity(self.d)
        self.sigma_mu=(1 / (self.beta + self.N * self.__exp_tau())) * I
        
        #Update mean
        sum_tn_w_xn = np.zeros([self.d,1]) #fix
        sum_tn_Wzn = np.sum((X - np.dot(self.__exp_W(), self.__exp_z())), axis=1, keepdims=True)
        self.mean_mu = self.__exp_tau() * np.dot(self.sigma_mu, sum_tn_Wzn)
       
    def __update_w(self, X):
        #Update sigma
        diag_alpha = np.diag(self.__exp_alpha())
        self.sigma_w = np.linalg.inv(np.add(np.diag(self.__exp_alpha()), self.__exp_tau()*self.__exp_z_zt()));
        
        #Update mean
        sum_xn_tnk_muk = np.subtract(X.T,self.mean_mu.T)
        self.means_w = self.__exp_tau() * np.dot(self.sigma_w, np.dot(self.means_z,sum_xn_tnk_muk)).T
        
    
    def __update_alpha(self):
        #Update a
        self.a_alpha_tilde = self.a_alpha + self.d/2
        
        #Update bs
        wi2 = np.add(np.transpose(self.__exp_W()*self.__exp_W()), np.diag(self.sigma_w))
        exp_norm_wi2 = np.sum( np.transpose(wi2), axis=0,keepdims=True)
        self.bs_alpha_tilde = self.b_alpha + 0.5*exp_norm_wi2

    def __update_tau(self, X):
        #Update a
        self.a_tau_tilde = np.array(self.a_tau + self.N*self.d/2)
        
        #Update b
        exp_norm_mu2 = np.sum(np.add(self.mean_mu*self.mean_mu,np.diag(self.sigma_mu)))
        norm_tn2 = np.sum(np.abs(np.transpose(X))*np.abs(np.transpose(X)),axis=None)
        sum_z_mu = (self.N * exp_norm_mu2 + norm_tn2);
        tr =  np.trace( np.dot(self.__exp_w_T_w(),( self.__exp_z_zt() )))
        mu_W_z = np.sum(np.dot( np.transpose(self.__exp_mu()) , np.dot( self.__exp_W() , self.__exp_z())))
        tt_W_z = np.sum(np.dot(np.transpose(X), np.dot(self.__exp_W(), self.__exp_z())))
        #tt_mu = np.sum(np.transpose(X), self.__exp_mu())
        tt_mu = np.sum(inner1d(np.dot(np.transpose(X),self.__exp_W()),np.transpose(self.__exp_z())))
        
        self.b_tau_tilde = self.b_tau + (1/2)*sum_z_mu + tr + 2*mu_W_z - 2*tt_W_z - 2*tt_mu

    def L(self, X):
        '''
        H_Q_Z = (self.d/2) *(1 + np.log(2*pi)) + (1/2)*np.log(np.linalg.det(self.sigma_z));
        H_Q_mu = (self.d/2) *(1 + np.log(2*pi)) + (1/2)*np.log(np.linalg.det(self.sigma_mu));
        H_Q_W = (self.d/2) *(1 + np.log(2*pi)) + (1/2)*np.log(np.linalg.det(self.sigma_w));
        H_Q_alpha = self.a_alpha_tilde - np.sum(map(lambda x: np.log(x), self.bs_alpha_tilde)) \
                    + np.log(sp.special.gamma(self.a_alpha_tilde))
                    +(1-self.a_alpha_tilde)*sp.special.digamma(self.a_alpha_tilde)
        H_Q_tau = self.a_tau_tilde - np.log(self.b_tau_tilde) \
                    + np.log(sp.special.gamma(self.a_tau_tilde))
                    +(1-self.a_tau_tilde)*sp.special.digamma(self.a_tau_tilde)
        '''
        #component1 = (self.d/2) * 
        
        L = 0.0
        return L
    
    def fit(self, X):
        for i in range(2000):
            self.__update_w(X)
            self.__update_z(X)
            self.__update_tau(X)
            self.__update_mu(X)
            self.__update_alpha()
```
